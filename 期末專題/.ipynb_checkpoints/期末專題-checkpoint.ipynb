```python
import tkinter as tk
import requests

keyfile = open("key.txt", "r")
key = keyfile.readline()
key = keyfile.readline()

api_key = key
location = "25.013413,121.540475"
keyword = "咖啡廳"

def search_nearby_places():

    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    params = {
        "key": api_key,
        "location": location,
        "radius": "2000",
        "keyword": keyword,
        "language": "zh-TW"
    }

    response = requests.get(url, params=params)
    data = response.json()

    results_listbox.delete(0, tk.END) # 刪除上一次的結果

    if data["status"] == "OK":
        results = data["results"]
        for place in results:
            name = place["name"]
            address_components = place["plus_code"]["compound_code"].split()
            county = [c for c in address_components if "縣" in c or "市" in c][0]
            address = place["vicinity"].replace(county, "")
            results_listbox.insert(tk.END, f"Name: {name}\n")
            results_listbox.insert(tk.END, f"Address: {county}{address}\n")
            results_listbox.insert(tk.END, "\n")
    else:
        results_listbox.insert(tk.END, "No results found.")

# 建立使用者介面
root = tk.Tk()
root.title("Nearby Places Search")

# 建立位置和關鍵字輸入框
location_label = tk.Label(root, text="Location:")
location_label.grid(row=0, column=0, padx=5, pady=5)
location_entry = tk.Entry(root, width=40)
location_entry.grid(row=0, column=1, padx=5, pady=5)
location_entry.insert(0, "National Taiwan Normal University")

keyword_label = tk.Label(root, text="Keyword:")
keyword_label.grid(row=1, column=0, padx=5, pady=5)
keyword_entry = tk.Entry(root, width=40)
keyword_entry.grid(row=1, column=1, padx=5, pady=5)
keyword_entry.insert(0, "Coffee Shop")

# 建立搜尋按鈕
search_button = tk.Button(root, text="Search", command=search_nearby_places)
search_button.grid(row=2, column=0, columnspan=2, padx=5, pady=5)

# 建立結果列表框架
results_frame = tk.Frame(root)
results_frame.grid(row=3, column=0, columnspan=2, padx=5, pady=5)

# 建立結果列表
results_listbox = tk.Listbox(results_frame, width=60, height=15)
results_listbox.pack(side=tk.LEFT, fill=tk.BOTH)
results_scrollbar = tk.Scrollbar(results_frame, orient=tk.VERTICAL, command=results_listbox.yview)
results_scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
results_listbox.config(yscrollcommand=results_scrollbar.set)

# 啟動使用者介面
root.mainloop()
```
