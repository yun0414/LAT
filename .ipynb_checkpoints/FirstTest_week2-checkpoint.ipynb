```python
import pandas as pd

df = pd.read_csv('106_1_2_5.csv')

df.head()
```

```python
temp = df[['學校所在地別', '畢業生人數', '公立大學日間部人數']]
temp
```

```python
Type = temp.groupby(['學校所在地別'])['公立大學日間部人數'].sum().reset_index()
Type['year'] = 106
Type
```

```python
df107 = pd.read_csv('107_1_2_5.csv')
df108 = pd.read_csv('108_1_2_5.csv')
df109 = pd.read_csv('109_1_2_5.csv')
```

```python
new = df107[['學校所在地別', '公立大學日間部人數']].groupby(['學校所在地別'])['公立大學日間部人數'].sum().reset_index()
```

```python
new['year'] = 107
new
```

```python
frames = [Type, new]

result = pd.concat(frames)

result
```

```python
new1 = df108[['學校所在地別', '公立大學日間部人數']].groupby(['學校所在地別'])['公立大學日間部人數'].sum().reset_index()

new1['year'] = 108

new1
```

```python
new2 = df109[['學校所在地別', '公立大學日間部人數']].groupby(['學校所在地別'])['公立大學日間部人數'].sum().reset_index()

new2['year'] = 109

new2
```

```python
frames1 = [Type, new, new1, new2]

result = pd.concat(frames1)

result
```
